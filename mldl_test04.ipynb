##**테스트**

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

#0이면 레드와인, 1이면 화이트 와인
df = pd.read_csv("https://bit.ly/wine_csv_data")
df.head()
```

```python
data = df[['alcohol', 'sugar', 'pH']].to_numpy()
target = df['class'].to_numpy()
print(data.shape)
print(target.shape)
```

```python
from tensorflow.keras.models import Sequential #모델 몸통
from tensorflow.keras.layers import Dense #모델 층

model =Sequential() # 비어 있는 모델 생성
model.add(Dense(1, input_dim=3, activation="sigmoid")) #로지스틱 회귀 모델

# 빌드(compile) 하기 : 모델 설정 했으므로 모델 완성하기
model.compile(optimizer = "sgd" ,loss = "binary_crossentropy")
model.fit(data, target, epochs=2000, verbose=0)
```

```python
# 예측결과가 0.5이하면 0으로
# 0.5 초과면 1로 분류가 됨
model.predict([[9.5, 7.7, 3.22]])
```

##**실습 S2**

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import Precision, Recall

# 데이터 로드
cancer = load_breast_cancer()
feature = cancer.data
target = cancer.target

# 스케일링
ss = StandardScaler()
feature = ss.fit_transform(feature)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)

# 모델 정의
model = Sequential()
model.add(Dense(1, input_dim=30, activation='sigmoid'))

# 모델 컴파일
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy', Precision(), Recall()])

# 모델 학습
model.fit(X_train, y_train, epochs=1000, verbose=0)

# 모델 평가
loss, accuracy, precision, recall = model.evaluate(X_test, y_test, verbose=0)
print(f'Loss: {loss}, Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}')
```

##**실습 S3**

```python
#K-Fold 교차 검증
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import Precision, Recall
import numpy as np

# 데이터 로드 및 전처리
cancer = load_breast_cancer()
feature = cancer.data
target = cancer.target

# 스케일링
ss = StandardScaler()
feature = ss.fit_transform(feature)

# K-Fold 교차 검증 설정
kf = KFold(n_splits=5, shuffle=True, random_state=42)
all_loss = []
all_accuracy = []
all_precision = []
all_recall = []

for train_index, test_index in kf.split(feature):
    X_train, X_test = feature[train_index], feature[test_index]
    y_train, y_test = target[train_index], target[test_index]

    # 모델 정의
    model = Sequential()
    model.add(Dense(1, input_dim=30, activation='sigmoid'))

    # 모델 컴파일
    model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy', Precision(), Recall()])

    # 모델 학습
    model.fit(X_train, y_train, epochs=1000, verbose=0)

    # 모델 평가
    loss, accuracy, precision, recall = model.evaluate(X_test, y_test, verbose=0)
    all_loss.append(loss)
    all_accuracy.append(accuracy)
    all_precision.append(precision)
    all_recall.append(recall)

# K-Fold 교차 검증 결과 요약
print(f'Average Loss: {np.mean(all_loss)}')
print(f'Average Accuracy: {np.mean(all_accuracy)}')
print(f'Average Precision: {np.mean(all_precision)}')
print(f'Average Recall: {np.mean(all_recall)}')
```
