```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

sn = np.genfromtxt('supernova.csv', delimiter=',', skip_header=1)
face = np.genfromtxt('face.csv', delimiter=',', skip_header=1)
box = np.genfromtxt('boxes.csv', delimiter=',', skip_header=1)
```

```python
print(sn)
print(face)
print(box)
```

```python
#supernova
# KMeans 알고리즘을 사용하여 군집화
kmeans = KMeans(n_clusters=4, random_state=35)  # 원하는 군집 수로 설정
kmeans.fit(sn)

# 군집화 결과
labels = kmeans.labels_

# 군집 중심
centers = kmeans.cluster_centers_

print("Cluster Labels:", labels)
print("Cluster Centers:", centers)
```

```python
# 만약 데이터가 2차원인 경우, 군집화 결과를 시각화할 수 있습니다.
plt.scatter(sn[:, 0], sn[:, 1], c=labels, cmap='Spectral')
plt.scatter(centers[:, 0], centers[:, 1])  # 군집 중심 표시

plt.title('supernova')
plt.show()
```

```python
# 최적의 k 찾기

inertia = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, n_init='auto', random_state=42)
    km.fit(sn)
    inertia.append(km.inertia_)

plt.plot(range(2, 7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```

```python
#face
kmeans = KMeans(n_clusters=4, random_state=1)  # 원하는 군집 수로 설정
kmeans.fit(face)

# 군집화 결과
labels = kmeans.labels_

# 군집 중심
centers = kmeans.cluster_centers_

print("Cluster Labels:", labels)
print("Cluster Centers:", centers)
```

```python
# 만약 데이터가 2차원인 경우, 군집화 결과를 시각화할 수 있습니다.
plt.scatter(face[:, 0], face[:, 1], c=labels, cmap='Spectral')
plt.scatter(centers[:, 0], centers[:, 1])  # 군집 중심 표시

plt.title('face')
plt.show()
```

```python
# 최적의 k 찾기

inertia = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, n_init='auto', random_state=42)
    km.fit(face)
    inertia.append(km.inertia_)

plt.plot(range(2, 7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
# k 값이 3에 근접할 때, 엘보우 포인트 (최적값)이 형성됨.
```
